```python
# Initialize workspace
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as pyp
import pandas as pd
from scipy import stats

# configure plot behavior
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x= np.arange(0,30,0.01)
fig, ax = pyp.subplots()
fig.set_size_inches((12,3))

ax.fill_between(x, stats.gumbel_l(7.5, 1).pdf(x), color = 'red');
ax.fill_between(x, stats.norm.pdf(x, 15, 1));
ax.fill_between(x, stats.gumbel_r(30-7.5, 1).pdf(x), color = 'green');
ax.axis('off');
```
